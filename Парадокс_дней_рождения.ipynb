```python
import random
```

```python
def paradox(max_people, num_runs=2000):
  for people in range(2, max_people + 1):
    # счетчик совпадений
    found_shared = 0
    for run in range(num_runs):
      bdays = []
      # генерация дня рождения для каждого человека в группе
      for i in range(0, people):
        bday = random.randrange(0, 365)
        bdays.append(bday)
      set_of_bdays = set(bdays) 

      # проверка на совпадение
      if len(set_of_bdays) < len(bdays):
        found_shared += 1

  prob = found_shared / num_runs
  return prob        
```

```python
print(paradox(23))
print(paradox(50))
```
